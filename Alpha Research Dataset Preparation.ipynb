```python
from tqdm import tqdm

sectors = []

for symbol in tqdm(nifty_500['Symbol_Yahoo'].values) :
    ticker = yf.Ticker(symbol)
    df = ticker.history(start = start,end = end,actions = False)
    
    df.columns = ['open','high','low','close','volume']
    
    if df.shape[0] < z.shape[0] :
        temp1 = []
        li1 = df.index.tolist()
        li2 = z.index.tolist()
        li_dif = [i for i in li1 + li2 if i not in li1 or i not in li2] 
        for date in li_dif:
            temp1.append([0,0,0,0,0])
        temp_df_1 = pd.DataFrame(temp1,index = li_dif,columns = df.columns)
        df = pd.concat([df,temp_df_1])
        df = df.sort_index(axis = 0)
        
    temp = []
    li1 = df.index.tolist()
    li2 = z.index.tolist()
    li_dif_1 = [i for i in li1 + li2 if i not in li1 or i not in li2] 
    df = df.drop(labels = li_dif_1,axis = 0)
    
    temp.append(list(df.loc['2009-12-22'].values))
    temp.append(list(df.loc['2009-12-22'].values))
    temp.append(list(df.loc['2009-12-22'].values))
    temp.append(list(df.loc['2009-12-22'].values))
    temp.append(list(df.loc['2009-12-22'].values))
    temp.append(list(df.loc['2010-11-04'].values))
    temp.append(list(df.loc['2011-10-25'].values))
    temp.append(list(df.loc['2012-10-25'].values))
    temp.append(list(df.loc['2012-11-12'].values))
    temp.append(list(df.loc['2014-10-22'].values))
    temp.append(list(df.loc['2015-11-10'].values))
    temp.append(list(df.loc['2019-02-12'].values))
    temp.append(list(df.loc['2019-03-28'].values))

    temp_df = pd.DataFrame(temp,index = index,columns = df.columns)

    df = pd.concat([df,temp_df])

    df = df.sort_index(axis = 0)
    
    df = df.drop(labels = [pd.Timestamp('2014-04-24'),pd.Timestamp('2014-10-15'),pd.Timestamp('2009-10-13')],axis = 0)
    
    df = df.fillna(method = 'ffill')
    
    df = df.loc[~df.index.duplicated(keep='first')]
    
    df.to_csv('data/daily/{}.csv'.format(nifty_500[nifty_500['Symbol_Yahoo'] == symbol]['Symbol'].values[0]))
    
```
